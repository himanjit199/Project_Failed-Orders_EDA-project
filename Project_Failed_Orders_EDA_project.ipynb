```python
a.drop_duplicates()
```

```python
a['cancellation_time_releated'] = a['m_order_eta']-a['cancellations_time_in_seconds']
```

```python
a.head()
```

```python
a['cancellation_time_releated_new']=np.where(a['cancellation_time_releated']>0,1,0)
```

```python
a.head()
a.drop('cancellation_time_releated',axis=1)
```

```python
def cancel_order(row):
    if row['order_status_key'] == 4:
        if row['is_driver_assigned_key'] == 0:
            if row['cancellation_time_releated_new'] == 0:
                return 'Cancelled by client before driver assignment and order is canceled before arrival'
            else:
                return 'Cancelled by client before driver assignment and order is canceled after arrival'
        else:
            if row['cancellation_time_releated_new'] == 0:
                return 'Cancelled by client after driver assignment and order is canceled before arrival'
            else:
                return 'Cancelled by client after driver assignment and order is canceled after arrival'
    elif row['order_status_key'] == 9:
        if row['cancellation_time_releated_new'] == 0:
            return 'Cancelled by system and order is canceled before arrival'
        else:
            return 'Cancelled by system and order is canceled after arrival'
    else:
        if row['cancellation_time_releated_new'] == 0:
            return 'Other reason and order is canceled before arrival'
        else:
            return 'Other reason and order is canceled after arrival'

# Applying the function to the DataFrame
a['failure_reason'] = a.apply(cancel_order, axis=1)






```

```python
a.head()
```

```python
c=a.failure_reason.value_counts().to_frame('count').reset_index()
c
```

```python
plt.figure(figsize=(40, 19))
sns.barplot(data=c,x='failure_reason',y='count')
plt.title('Distribution of Orders According to Reasons for Failure', fontsize=20)
plt.xlabel('Failure Reason', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
```

Hence **"Cancelled by client before driver assignment and order is canceled after arrival"** has highest cancelation number .

```python
e=pd.read_csv('data_orders.csv')
e.head()
```

```python
e['cancellations_time_in_seconds'] =e['cancellations_time_in_seconds'].fillna(e.cancellations_time_in_seconds.mean())
```

```python
e['order_datetime'] = pd.to_datetime(e['order_datetime'])
e['hours']=e['order_datetime'].dt.hour
```

```python
g=e.drop(e.iloc[:,:-4],axis=1)
g.head()
```

```python
def interval(rows):
    hour = rows['hours']
    if 1 <= hour <= 8:
        return "between 1AM to 8AM"
    elif 8 < hour <= 16:
        return "after 8AM to 4PM"
    elif 16 < hour <= 21:
        return "after 4PM to 9PM"
    else:
        return "after 9PM to 12AM"
```

```python
g['time_interval']=g.apply(interval,axis=1)
```

```python
g.head()
```

```python
average_time_hour =g.groupby(['is_driver_assigned_key','time_interval'])['cancellations_time_in_seconds'].mean().reset_index()
average_time_hour
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=average_time_hour,x='time_interval',y='cancellations_time_in_seconds',hue='is_driver_assigned_key')
plt.title('Average Cancellation Time by Hour and Driver Assignment')
```

Final Results are :

1.    "Cancelled by client before driver assignment and order is canceled after arrival" has highest cancelation number
2.   between *1 am to 8 Am * highest failure of order happens . ❎

3. Highest Average ETA is seen in  8 Am interval.
4. "after 9PM to 12AM" , in this interval average cancellation time is highest with driver or without driver in both case.
